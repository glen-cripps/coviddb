```python
import requests 
import pandas as pd 
from bs4 import BeautifulSoup

import arrow
today_dt = arrow.now().format('YYYYMMDD')
```

```python
yahoo_tickers=[]
names=[]
prices=[]
price_changes=[]
price_changes_pct=[]
volumes = []

CryptoCurrenciesUrl = "https://www.finance.yahoo.com/world-indices"
r= requests.get(CryptoCurrenciesUrl)
data=r.text
soup=BeautifulSoup(data)

for i in range(46, 641, 17):
    for listing in soup.find_all('tr', attrs={'data-reactid':i}):
        for yahoo_ticker in listing.find_all('td', attrs={'data-reactid':i+1}):
            yahoo_tickers.append(yahoo_ticker.text)
        for name in listing.find_all('td', attrs={'data-reactid':i+3}):
            names.append(name.text)
        for price in listing.find_all('td', attrs={'data-reactid':i+4}):
            prices.append(price.text)
        for price_change in listing.find_all('td', attrs={'data-reactid':i+5}):
            price_changes.append(price_change.text)
        for price_change_pct in listing.find_all('td', attrs={'data-reactid':i+7}):
            price_changes_pct.append(price_change_pct.text)
        for volume in listing.find_all('td', attrs={'data-reactid':i+9}):
            volumes.append(volume.text)
        

#for i in range(46, 47, 14):

df = pd.DataFrame({"yahoo_ticker":yahoo_tickers, "names": names, "price": prices, "price_change": price_changes, "price_change_pct":price_changes_pct, "volume":volumes })
df
```

```python
df.to_parquet("mwis." + today_dt + ".parquet", compression='GZIP')
```
