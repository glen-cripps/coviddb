```python
import datetime
num_days=7
start_sp = datetime.date.today() - datetime.timedelta(days=num_days)
end_sp = datetime.date.today()
from pandas_datareader import data as pdr
import pandas as pd

import arrow
today_dt = arrow.now().format('YYYYMMDD')
financial_indexes = pd.read_parquet("mwis." + today_dt + ".parquet")
```

```python
# initialize a dataframe with the columns we want
all_financials_df = pd.DataFrame()
all_financials_df = pdr.get_data_yahoo('^GSPC', start_sp, end_sp)
all_financials_df['yahoo_ticker'] = '^GSPC'

# then delete the rows coz it will reload them in the loop below
all_financials_df = all_financials_df[0:0]

for fin_index in financial_indexes['yahoo_ticker']:
    try:
        financials_df = pdr.get_data_yahoo(fin_index, start_sp, end_sp)
        financials_df['yahoo_ticker'] = fin_index
        financials_df['date'] = financials_df.index
    except:
        continue
    all_financials_df = all_financials_df.append(financials_df)
```

```python
all_financials_df = all_financials_df.set_index('close').reset_index()
all_financials_df = all_financials_df.set_index('date').reset_index()
all_financials_df = all_financials_df.set_index('yahoo_ticker').reset_index()


all_financials_df.columns = all_financials_df.columns.str.lower()
all_financials_df.columns = all_financials_df.columns.str.replace(' ', '_')
all_financials_df.columns = all_financials_df.columns.astype(str)
```
